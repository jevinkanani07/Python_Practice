#### 7. write a python programm to check if a number is prime.

```python
a = int(input("Enter your number to check prime number :-"))

if a <= 1:
    print(a,"is not prime number ")
else:
    for i in range(2,a):
        if a % i == 0:
            print(a,"is not prime number ")
            break

    else:
        print(a,"is a prime number")
```
